### Print the label noise transition matrix before and after score curation for Flan\_v2 dataset

```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import torch

# dataset_name ='oasst1'
dataset_list = ['utilitarian']

for dataset_name in dataset_list:

    report_path = f"score_curation_results/{dataset_name}_report.pt"
    reports = torch.load(report_path)


    data = reports.diagnose['T']

    plt.figure(figsize=(8, 6))

    sns.heatmap(data, annot=True, fmt=".2f", cmap="YlGnBu")

    # plt.title(f'Label transition matrix: {dataset_name}; Labeling model: {model_name}', fontsize=18)
    plt.title(f'Score transition matrix ({dataset_name})', fontsize=18)
    plt.xlabel('Scores', fontsize=18)
    plt.ylabel('Scores', fontsize=18)

    
    plt.show()
```
